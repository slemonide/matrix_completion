```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot4x4(A,B,C,D):
          # A     B     C     D
    fig, ((ax11, ax12, ax13, ax14),
          (ax21, ax22, ax23, ax24),
          (ax31, ax32, ax33, ax34),
          (ax41, ax42, ax43, ax44)) = plt.subplots(4, 4, sharex='col', sharey='row')

    # First row
    ax11.plot(A, A, '*')
    ax11.set_title("A")
    ax11.set_ylabel("A")

    ax12.plot(A, B, '*')
    ax12.set_title("B")

    ax13.plot(A, C, '*')
    ax13.set_title("C")

    ax14.plot(A, D, '*')
    ax14.set_title("D")

    # Second row
    ax21.plot(B, A, '*')
    ax21.set_ylabel("B")
    ax22.plot(B, B, '*')
    ax23.plot(B, C, '*')
    ax24.plot(B, D, '*')

    # Third row
    ax31.plot(C, A, '*')
    ax31.set_ylabel("C")
    ax32.plot(C, B, '*')
    ax33.plot(C, C, '*')
    ax34.plot(C, D, '*')

    # Fourth row
    ax41.plot(D, A, '*')
    ax41.set_ylabel("D")
    ax42.plot(D, B, '*')
    ax43.plot(D, C, '*')
    ax44.plot(D, D, '*')
```

```python
a, b, c, d, ph, th = symbols('a b c d phi theta')
```

```python
s1, s2 = symbols('s1 s2')
```

```python
U = Matrix([[a, b], [-exp(I*ph)*conjugate(b), exp(I*ph)*conjugate(a)]])
```

```python
V = Matrix([[c, d], [-exp(I*th)*conjugate(d), exp(I*th)*conjugate(c)]])
```

```python
E = Matrix([[s1, 0], [0, s2]])
```

```python
X = U*E*transpose(conjugate(V))
```

```python
P = Matrix([[1, 0], [0, 0]])
```

```python
Y = X - P
```

```python
Y
```

```python
det(Y)
```

```python
Y.subs([(a, 1), (b, 0), (c, 1), (d, 0), (ph, 0), (th, 0), (s1, 1), (s2, 0)])
```

```python
def plotfor(thv,phv,sign1,sign2):
    
    A = []
    B = []
    C = []
    D = []

    SMPL = 10
    for i in range(0,SMPL):
        av = np.sin(2*np.pi*i/SMPL)
        bv = sign1*np.cos(2*np.pi*i/SMPL)
        for j in range(0,SMPL):
            cv = np.sin(2*np.pi*j/SMPL)
            dv = sign2*np.cos(2*np.pi*j/SMPL)
            for s1i in range(0,SMPL):
                s1v = s1i/SMPL
                s2v = 1 - s1v

                M = np.array(Y.subs([(a, av),
                                     (b, bv),
                                     (c, cv),
                                     (d, dv),
                                     (ph, phv),
                                     (th, thv),
                                     (s1, s1v),
                                     (s2, s2v)]).as_real_imag()[0]).astype(np.float64)
                A.append(M[0,0])
                B.append(M[0,1])
                C.append(M[1,0])
                D.append(M[1,1])

    plot4x4(A,B,C,D)
```

```python
sign1 = 1
sign2 = 1
plotfor(0,0,sign1,sign2)
plotfor(np.pi,0,sign1,sign2)
plotfor(0,np.pi,sign1,sign2)
plotfor(np.pi,np.pi,sign1,sign2)
```

```python
sign1 = -1
sign2 = 1
plotfor(0,0,sign1,sign2)
plotfor(np.pi,0,sign1,sign2)
plotfor(0,np.pi,sign1,sign2)
plotfor(np.pi,np.pi,sign1,sign2)
```

```python
sign1 = 1
sign2 = -1
plotfor(0,0,sign1,sign2)
plotfor(np.pi,0,sign1,sign2)
plotfor(0,np.pi,sign1,sign2)
plotfor(np.pi,np.pi,sign1,sign2)
```
