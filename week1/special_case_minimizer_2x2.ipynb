```python
x, y = symbols('x y')
```

```python
M = Matrix([[1, y], [x, 0]])
```

```python
M
```

For these kind of matrices, singular values can be computed explicitly:

```python
S1 = 1 + x**2 + y**2
S2 = sqrt((1 + y**2 - x**2)**2 + 4*x**2)

sigma1 = sqrt((S1 + S2)/2)
sigma2 = sqrt((S1 - S2)/2)

norm = sigma1 + sigma2

norm
```

To find minimum, compute gradient:

```python
gradient = Matrix([norm.diff(x), norm.diff(y)])
gradient
```

```python
extremas = solve(gradient,x,y)
```

```python
plot3d(norm, (x, -0.0001, 0.0001), (y, -0.0001, 0.0001))
plot3d(norm, (x, -0.01, 0.01), (y, -0.01, 0.01))
```
